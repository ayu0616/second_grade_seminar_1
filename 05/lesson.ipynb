この相関行列の固有値を計算すると主成分分析ができる

```r
okamoto = read.csv("./okamoto.csv")
```

丘本円周データは2次元で表せるはずの場所を5次元で表している  
主成分分析にかけると無駄を省けるのではないか

```r
oka.pc = princomp(okamoto)
```

```r
summary(oka.pc, digits = digits, maxsum = maxsum)
```

```r
# loadings = 負荷量
oka.pc$loadings
```

```r
# scores = 主成分得点
oka.pc$scores
```

```r
plot(oka.pc$scores[, 1:2])
```
