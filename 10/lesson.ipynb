# ロジスティック関数：$\frac{a}{1 + be^{cx}}$


下の式は $y=100x^3+（誤差）$

```r
set.seed(30)
x = seq(-5, 5, 0.1)
y <- 10 * x^3 + 100 * rnorm(x, 0, 1)
plot(x, y)
```

```r
tm3 = nls(y ~ a * x^3 + b * x^2 + c * x + d, start = c(a = 1, b = 1, c = 1, d = 1), trace = TRUE) # nolint
```

```r
summary(tm3)
```
