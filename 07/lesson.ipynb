```r
(kansai = read.csv("./kansai.csv", row.names = 1))
```

```r
kansai.cmd = cmdscale(kansai, k=2, eig=FALSE)
```

```r
par(family= "YuKyo_Yoko-Medium")
plot(kansai.cmd, type="n")
text(kansai.cmd, rownames(kansai.cmd))
```
