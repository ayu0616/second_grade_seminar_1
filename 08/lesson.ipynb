```r
seiseki = read.csv("./seiseki.csv", row.names = 1 ,fileEncoding="utf8")
seiseki
```

```r
seiseki.d = dist(seiseki)
```

```r
seiseki.hc = hclust(d = seiseki.d)
```

```r
par(family = "kyokasho")
plot(seiseki.hc)
```

```r
par(family = "YuKyo_Yoko-Medium")
plot(hclust(seiseki.d, method = "single"))
```

```r
par(family = "YuKyo_Yoko-Medium")
plot(hclust(seiseki.d, method = "centroid"))
```

```r
par(family = "YuKyo_Yoko-Medium")
plot(hclust(seiseki.d, method = "ward.D"))
```

```r
par(family = "kyokasho")
heatmap(as.matrix(seiseki))
```
